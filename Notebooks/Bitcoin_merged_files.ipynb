```python
def load_single_output_file(download_path, file_name):
    # Define the folder for transactions
    trx_folder = os.path.join(download_path, 'outputs')

    # Full path to the transaction file
    file_path = os.path.join(trx_folder, file_name)

    # Check if the file exists
    if os.path.exists(file_path) and file_path.endswith(".tsv.gz"):
        # Load the transaction file into a dataframe
        df_trx = pd.read_csv(file_path, compression='gzip', sep='\t')
        return df_trx
    else:
        print(f"File {file_name} does not exist in the inputs folder.")
        return None
```

```python
def load_single_input_file(download_path, file_name):
    # Define the folder for transactions
    trx_folder = os.path.join(download_path, 'inputs')

    # Full path to the transaction file
    file_path = os.path.join(trx_folder, file_name)

    # Check if the file exists
    if os.path.exists(file_path) and file_path.endswith(".tsv.gz"):
        # Load the transaction file into a dataframe
        df_trx = pd.read_csv(file_path, compression='gzip', sep='\t')
        return df_trx
    else:
        print(f"File {file_name} does not exist in the inputs folder.")
        return None
```

```python
def load_single_transaction_file(download_path, file_name):
    # Define the folder for transactions
    trx_folder = os.path.join(download_path, 'transactions')

    # Full path to the transaction file
    file_path = os.path.join(trx_folder, file_name)

    # Check if the file exists
    if os.path.exists(file_path) and file_path.endswith(".tsv.gz"):
        # Load the transaction file into a dataframe
        df_trx = pd.read_csv(file_path, compression='gzip', sep='\t')
        return df_trx
    else:
        print(f"File {file_name} does not exist in the transactions folder.")
        return None
```

```python
def load_dataframes(download_path):
    # Define the folders for transactions and inputs
    trx_folder = os.path.join(download_path, 'transactions')
    input_folder = os.path.join(download_path, 'inputs')

    # Initialize empty dataframes for transactions and inputs
    df_trx_all = pd.DataFrame()
    df_input_all = pd.DataFrame()

    # Load all transaction files into a single dataframe
    for file in os.listdir(trx_folder):
        file_path = os.path.join(trx_folder, file)
        if file.endswith(".tsv.gz"):
            df_trx = pd.read_csv(file_path, compression='gzip', sep='\t')
            df_trx_all = pd.concat([df_trx_all, df_trx], ignore_index=True)

    # Load all input files into a single dataframe
    for file in os.listdir(input_folder):
        file_path = os.path.join(input_folder, file)
        if file.endswith(".tsv.gz"):
            df_input = pd.read_csv(file_path, compression='gzip', sep='\t')
            df_input_all = pd.concat([df_input_all, df_input], ignore_index=True)

    return df_trx_all, df_input_all
```

```python
import os
import pandas as pd
from datetime import datetime

def load_account_input_files(download_path):
    # Define the folder for input account details
    folder = os.path.join(download_path, 'input acc details csv')

    # Initialize an empty dataframe for all data
    df_all = pd.DataFrame()

    # Get all file names in the folder
    files = os.listdir(folder)

    # Filter out only the relevant files and extract the date part from the file name
    filtered_files = [(file, datetime.strptime(file.split('_')[-1].split('.')[0], '%Y%m%d'))
                      for file in files if file.endswith(".tsv.gz")]

    # Sort files by the extracted date
    sorted_files = sorted(filtered_files, key=lambda x: x[1])

    # Load and concatenate the files in sorted order
    count = 0
    for file, _ in sorted_files:
        file_path = os.path.join(folder, file)
        count += 1
        df = pd.read_csv(file_path, compression='gzip', sep='\t')
        df_all = pd.concat([df_all, df], ignore_index=True)

    print(count)

    return df_all
```

```python
def load_account_output_files(download_path):
    # Define the folders for transactions and inputs
    folder = os.path.join(download_path, 'output acc details csv')

    # Initialize empty dataframes for transactions and inputs
    df_all = pd.DataFrame()

    # Load all transaction files into a single dataframe
    for file in os.listdir(folder):
        file_path = os.path.join(folder, file)
        if file.endswith(".tsv.gz"):
            df = pd.read_csv(file_path, compression='gzip', sep='\t')
            df_all = pd.concat([df_all, df], ignore_index=True)

    return df_all
```
